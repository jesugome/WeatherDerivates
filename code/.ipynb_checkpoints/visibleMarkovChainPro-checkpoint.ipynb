### Precipitation Amount Fit

```python
import scipy.stats as stats    

gammaAlpha, gammaLoc, gammaBeta=stats.gamma.fit(noZeroPrecipitation['Prep'])

fittedParameter = {
    'Alpha': [gammaAlpha],
    'Loc': [gammaLoc],
    'Beta': [gammaBeta] 
}

fittedGamma = pd.DataFrame(fittedParameter)
fittedGamma
```

```python
# Plot fitted gamma
r = gamma.rvs(gammaAlpha, loc = gammaLoc, scale = gammaBeta, size=10000)
fig = plt.figure(figsize=(20, 10))
plt.hist(r,color='violet',bins=100)
plt.title('Fitted Data')
plt.show()
```

### Transition Matrix

```python
# Build transition matrix entries.

# Find the count of wet days that followed dry days.

dryDays = list(np.where(precipitationAllTime['Prep']==0.0)[0])

n01=0

for i in dryDays:
    if i == dryDays[-1]:
        break
    if precipitationAllTime.iloc[i+1,0] > 0:
        n01+=1
```

```python
# Find the count of dry days that followed dry days.

dryDays = list(np.where(precipitationAllTime['Prep']==0.0)[0])

n00=0

for i in dryDays:
    if i == dryDays[-1]:
        break
    if precipitationAllTime.iloc[i+1,0] == 0:
        n00+=1
```

```python
# Find the count of wet days that followed wet days.

wetDays = list(np.where(precipitationAllTime['Prep']>0.0)[0])

n11=0

for i in wetDays:
    if i == wetDays[-1]:
        break
    if precipitationAllTime.iloc[i+1,0] > 0:
        n11+=1
```

```python
# Find the count of dry days that followed dry days.

wetDays = list(np.where(precipitationAllTime['Prep']>0.0)[0])

n10=0

for i in wetDays:
    if i == wetDays[-1]:
        break
    if precipitationAllTime.iloc[i+1,0] == 0:
        n10+=1
```

```python
# Build probabilities using counts. 
p01 = n01/(n00+n01)
p11 = n11/(n10+n11)

p00 = 1- p01
p10 = 1- p11

# Print transition 
transitionMatrix = pd.DataFrame({'0':[p00,p10],'1':[p01,p11]})

print('Transition Matrix applied to rainfall data is: \n 0 == Dry \n 1 == Wet ')
transitionMatrix
```

```python
ohhh =[40,50]
transitionMatrix['1'] = ohhh
```

```python
def findTransitions(iniState,finState,dataframe):
    iniStateDays = list(np.where(dataframe['State']==iniState)[0])
    
    transitionsTofin = 0

    for i in iniStateDays:
        if i == iniStateDays[-1]:
            break
        if dataframe['State'][i+1] == finState:
            transitionsTofin+=1
    
    return transitionsTofin
```

## Seasonality Analysis

```python
startYear = 1972
endYear = 2015

precipitationWindowTime=precipitationAllTime.loc[str(startYear)+'-01-01':str(endYear)+'-12-31']
```

```python
# Extract month from index.

indexTime = precipitationWindowTime.index
precipitationWindowTime['Month']= indexTime.month
precipitationWindowTime['State'] = np.where(precipitationWindowTime['Prep']>0,1,0)
precipitationWindowTime.head()
```

```python

# Find the number of transitions for all possible transitions.  

def monthTransitionsCalculator(numStates,data):
    
    transitions = pd.DataFrame(columns=['n00','n01','n10','n11'])
    
    for month in range(1,12+1):
        dataframe = data[data['Month']==month]
        
        index = 0

        transitionProb=[0]*numStates*numStates

        for i in range(numStates):
            for j in range(numStates):
                transitionProb[index] = findTransitions(i,j,dataframe)
                index+=1

        transitions.loc[month] = transitionProb
    return transitions    
```

```python
# Print Transitions for each month. 
print('Number of Transition applied to rainfall data is: \n 0 == Dry \n 1 == Wet ')
monthTransitions = monthTransitionsCalculator(2,precipitationWindowTime)
monthTransitions.index.name='Month'
monthTransitions
```

```python
def withTransitionsGetProb(monthTransitions):
    
    monthTransitionProb = pd.DataFrame(columns=['p00','p01','p10','p11'])
    
    probs=[0]*4
    
    for month in range(1,13):
        
        # Build probabilities using counts. 
        p01 = monthTransitions.loc[month]['n01']/(monthTransitions.loc[month]['n00']+monthTransitions.loc[month]['n01'])
        p11 = monthTransitions.loc[month]['n11']/(monthTransitions.loc[month]['n10']+monthTransitions.loc[month]['n11'])

        p00 = 1- p01
        p10 = 1- p11
        
        probs = [p00,p01,p10,p11]
        
        monthTransitionProb.loc[month] = probs
    return monthTransitionProb
        
```

```python
monthTransitionsProb = withTransitionsGetProb(monthTransitions)
monthTransitionsProb
```

```python
plot = monthTransitionsProb.plot(figsize=(20,10), title = 'Transition Probabities',kind='bar',grid=True)
plot.set_xlabel('Month')
plot.set_ylabel('Probability')
```
